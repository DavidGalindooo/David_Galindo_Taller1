# A view on pespair
##By Sonja Kuijpers at studio terp

En la siguiente visualización nos brindan un panorama donde nos representan con arte el número
de suicidios en los países bajos en el año 2017. Proyecto formado a raíz de una lucha personal
contra la depresión.

##WHAT? WHY? HOW?

###WHAT?
En esta visualización podemos admirar un conjunto de atributos en campo de tipo categóricos (elementos representativos)
divergentes que nos indican la cantidad y el método de suicidio. Árboles para estrangulamiento,
olas para representan ahogo, las nubes muerte por alcohol, drogas y medicamentos, las luces
encima de las montañas representan suicidio frente al metro/por el metro, los edificios son
suicidio por salto de altura, las hojas en la parte inferior representan otro método no mencionado
anteriormente, y por último en la parte superior, las estrellas representan muerte con métodos
desconocidos.

##WHY?
Representar suicidios, concientizar acerca de la cantidad de suicidios en los países bajos. Al no ser
una visualización interactiva, la única labor que tiene el espectador es admirarla y analizarla con
atención en cada uno de sus elementos.

##HOW?
Se usa una codificación tipo mapa que nos brinda información a partir de formas que contienen
colores, saturación iluminación y demás.

##MARCAS Y CANALES
Se usan marcas de áreas con formas familiares donde se usan canales de posición espacial tanto horizontal como vertical y formas de alta relevancia

Se cumplen las reglas generales para cumplir con una buena visualización de datos.

##Mejoras posibles:


*   Brindarle animación y saturación para indicar un mayor o menor índice de muertes por
cada mes/semana del año.
*   Separar hombres y mujeres por formas y/o colores.

#Codex atlanticus
##By the visual agency


En la siguiente visualización nos brindan un vistazo al Codex Atlanticus de Leonardo Da Vinci. Un proyecto **interactivo** que abre nuevas alternativas de explorar el pensamiento de Leonardo. Una amplia vista panorámica con opción de filtro según temas de preferencias y reorganización cronológica del codex como tal .



a1.png

#WHAT? WHY? HOW?


##WHAT?
En esta visualización podemos admirar un conjunto de atributos en campo  de tipo ordinales que en un principio mantienen un orden secuencial, pero que nos brinda la posibilidad de admirar y reorganizar las páginas del Codex Atlanticus ( una recopilación de escritos y dibujos de Leonardo Da Vinci) según interés u orden cronológico.

##WHY?
Conservar el Codex Atlanticus como legado, facilitar el estudio del Codex Atlanticus. Las transformaciones requeridas como filtros y orden, al igual que datos particulares  dependen únicamente de las necesidades del espectador.


##HOW?
Se usa codificación de tipo mapa ordenado donde los colores nos indican categorías, y en el cual podemos realizar filtros los cuales resultan diferenciados por posición y color.
